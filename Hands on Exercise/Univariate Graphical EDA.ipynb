### Modules

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

### Pie Chart

```python
languages = ['Python', 'R', 'Java', 'HTML', 'C++']
popularity = [22.8, 16, 10, 8, 9]
colors = ['#FA8F0A', '#F0DA32', '#61F527', '#18DE9D', '#3594FF']

explode =(0.1, 0, 0, 0, 0)
plt.pie(x=popularity, labels=languages, colors=colors, autopct="%1.1f%%", explode=explode)
plt.title('Popularity of Programming Languages')
plt.show()
```

### Histogram - Vertical

```python
values = [82, 76, 24, 67, 40, 62, 75, 78, 71, 32, 98, 89, 78, 67, 72, 82, 87, 66, 56, 52, 65, 85]
plt.hist(values, bins=10, edgecolor='black', label='Test score data')
plt.title('Histogram for scores')
plt.xlabel('Scores')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

### Histogram - Horizontal

```python
values = [82, 76, 24, 67, 40, 62, 75, 78, 71, 32, 98, 89, 78, 67, 72, 82, 87, 66, 56, 52, 65, 85]
plt.hist(values, bins=10, edgecolor='black', label='Test score data', orientation='horizontal')
plt.title('Histogram for scores')
plt.xlabel('Scores')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

### Line Plot

```python
year = [1960, 1970, 1980, 1990, 2000, 2010]
pop_china = [667.07, 818.32, 981.24, 1135.20, 1262.60, 1337.70]
pop_india = [449.48, 553.58, 696.78, 870.13, 1053.10, 1231.00]
pop_usa = [180.67, 205.05, 227.23, 249.62, 282.162, 309.34]

plt.plot(year, pop_china, label='China', color='red')
plt.plot(year, pop_india, label='India', color='green')
plt.plot(year, pop_usa, label='USA', color='blue')

plt.xlabel('Country Names')
plt.ylabel('Population Data')
plt.legend()

plt.show()
```

### Importing dataset

```python
df = pd.read_csv("cancer.csv")
df.head()
```

### Box Plot - Vertical

```python
plt.figure(figsize=(10, 6))
plt.boxplot(df['area_mean'])
plt.show()
```

### Box Plot - Horizontal

```python
plt.figure(figsize=(10, 6))
plt.boxplot(df['area_mean'], vert=False)
plt.show()
```

### Grouping the data using Boxplot

```python
df.boxplot(column = 'area_mean', by='diagnosis')
plt.title('Grouping the data using boxplot')
plt.show()
```

### Boxplot using Seaborn

```python
sns.boxplot(data=df, x='diagnosis', y='area_mean')
plt.title('Boxplot using Seaborn library')
plt.show()
```
