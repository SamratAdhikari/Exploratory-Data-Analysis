### Modules

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

### Loading a dataset from seaborn library for scatter plot

```python
tips = sns.load_dataset('tips')
tips.head()
```

### Looking at Five Point Summary

```python
tips.describe()
```

### Scatter Plot

```python
plt.title('Scatter plot between Total Bill and Tip columns')
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex')
plt.show()
```

### Linear Regression line on Scatter plot

```python
sns.lmplot(data=tips, x='total_bill', y='tip', hue='sex', markers=['.', 'x'])
plt.show()
```

### Using two seperate scatter plots for each gender

```python
sns.lmplot(data=tips, x='total_bill', y='tip', col='sex')
plt.show()
```

### Loading dataset for heat map

```python
df = pd.read_csv('cars.csv')
df.head()
```

### Renaming the first column for better analysis

```python
df.columns
df.rename(columns={'Unnamed: 0': 'model'}, inplace=True)
df.head()
```

### Creating a Scatter Plot Matrix (Pair Plot)

```python
sns.pairplot(df)
plt.show()
```

### Creating a Heat Map

```python
# correlation value
df.corr()

plt.figure(figsize=(10, 10))
sns.heatmap(df.corr(), annot=True, cmap='BrBG')
plt.show()
```
