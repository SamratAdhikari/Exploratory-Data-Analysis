### Modules

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sbn

import warnings
warnings.filterwarnings('ignore')
```

### Initializing the dataframe

```python
df = pd.read_csv('covid19_Confirmed_dataset.csv')
df.head()
```

### Cleaning the datarame

```python
df.drop(columns=['Province/State', 'Lat', 'Long'], inplace=True)
df.head()
```

### Looking for any null values

```python
df.isnull().sum()
```

### Country wise data for each year

```python
aggregation = df.groupby('Country/Region').sum()
aggregation
```

```python
aggregation.shape
```

### Visualize the data for China, Italy & Spain

```python
aggregation.loc['China'].plot(color='red')
aggregation.loc['Italy'].plot(color='green')
aggregation.loc['Spain'].plot(color='purple')

plt.legend()
plt.show()
```

Spain & Italy are most affected by Covid19

### Maximum Infection Rate

```python
# China
aggregation.loc['China'].diff().max()

# Italy
aggregation.loc['Italy'].diff().max()

# Spain
aggregation.loc['Spain'].diff().max()
```

```python
countries = list(aggregation.index)
max_infection_rate = []

[max_infection_rate.append(aggregation.loc[x].diff().max()) for x in countries]

aggregation['max_infection_rate'] = max_infection_rate
aggregation.head()
```

### Get only the required data

```python
data = pd.DataFrame(aggregation['max_infection_rate'])
data.head()
```

### Initializing another dataframe

```python
happiness = pd.read_csv('worldwide_happiness_report.csv')
happiness.head()
```

```python
cols_to_drop = ['Overall rank', 'Score', 'Generosity', 'Perceptions of corruption']
happiness.drop(columns=cols_to_drop, inplace=True)

happiness.head()
```

### Setting 'Country or region' as index

```python
happiness.set_index('Country or region', inplace=True)
happiness.head()
```

### Concatinating the recent dataframe with 'data' dataframe

```python
final_data = data.join(happiness, how='inner')
final_data.head()
```

### Correlation matrix between variables

```python
sbn.heatmap(final_data.corr(), annot=True, cmap='Greens')
plt.show()
```

```python
x = final_data['GDP per capita']
y = final_data['max_infection_rate']

sbn.regplot(data=final_data, x=x, y=np.log(y))
plt.show()
```
